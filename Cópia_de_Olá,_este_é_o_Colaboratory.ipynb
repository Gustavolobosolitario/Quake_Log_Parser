##Importação das libs


```
import pandas as pd
import numpy as np
import json
```

#Importando dados

```
# dados de ingressos
df_ingressos = pd.read_json('https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_ingressos')

# dados dos shows
df_shows = pd.read_json('https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_shows')

# dados de compras
df_compras = pd.read_csv('https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_compras')
```

```
# Visualizando dados de ingressos
df_ingressos.head()
```

```
# Visualizando dados dos shows
df_shows.head()
```

```
# Visualizando dados de compras
df_compras.head()
```

```
# Pessoas Ingressos Pista
df_ingressos.loc[df_ingressos['tipo']=='Pista']
```

```
# União da tabela de ingressos e compras
df_ingressos_gastos = df_ingressos.merge(df_compras, on='nome', how = 'inner')
df_ingressos_gastos
```

```
# Média de gastos de pessoas da pista
df_ingressos_gastos[df_ingressos_gastos['tipo']=='Pista']['gastos'].mean()
```

```
# 3-Pessoas compraram ingressos com concorrentes
df_ingressos[df_ingressos['status']=='Nao Concluido']
```

```
# 4- O dia com maior gasto
df_compras.groupby('show')['gastos'].sum()
```

```
#5- clientes que desistiram de comprar o ingresso com a AT
df_ingressos[df_ingressos['status']=='Nao Concluido']
```
