# 4- Qual o dia com maior gasto?

* Foi filtrado a soma de gastos dos três dias.

```python
df_ingressos_gastos.groupby('dia')['gastos'].sum()
```

* Impresso o gasto total de cada dia em lista.

```python
df_gastos_por_dia = pd.DataFrame(df_ingressos_gastos.groupby('dia')['gastos'].sum())
```

```python
df_gastos_por_dia
```

* Foi feito um index para saber que o dia que obteve mais gastos foi dia 1.

```python
df_gastos_por_dia.index[df_gastos_por_dia.gastos==df_gastos_por_dia.gastos.max()].to_list()
```
